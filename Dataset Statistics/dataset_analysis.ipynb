```python
import pandas as pd

# Load the dataset
train_df = pd.read_csv('dataset/train.csv')

# Function to calculate the distribution of each category
def calculate_distribution(dataframe, column):
    count = dataframe[column].value_counts()
    percentage = dataframe[column].value_counts(normalize=True) * 100
    distribution = pd.DataFrame({'Count': count, 'Percentage': percentage})
    return distribution

# List of fine-grained categories to analyze
categories = ['Disability', 'Gender/Sexual', 'Origin/Nationality', 'Race/Ethnicity', 'Religion']

# Display the distribution for each category
for category in categories:
    print(f"Distribution for {category}:\n{calculate_distribution(train_df, category)}\n")
```
